```python
all_df = []

#read in csv files
for file in csv_files:
    df = pd.read_csv(file, index_col=0, header=0)
    all_df.append(df)
    #change column names
    df.columns = ['image_name', 'area', 'mean', 'stddev', 'intden', 'rawintden', 'length']
    df.rename(columns = {'length':'length_single_cohesin_axis'}, inplace = True)
    df.rename(columns = {'label':'image_name'}, inplace = True)
    df["condition"] = "single_cohesin_axis_wapl_depleted"

#concat all dataframes together
frame = pd.concat(all_df, axis=0, ignore_index=True)
#remove duplicated rows
cols = frame.columns
#dedup = frame[cols].loc[(frame[cols].shift() != frame[cols]).any(axis=1)]

#slice relevant columns into final dataframe
length_single_axis = frame[['image_name','length_single_cohesin_axis', 'condition']]

#fully_prometa_two_sister_segment = fully_prometa_two_sister_segment.sort_values('edu_segment_length', ascending=False)
#fully_prometa_two_sister_segment_reset = fully_prometa_two_sister_segment.reset_index(drop=True)

results_dir = "/groups/gerlich/experiments/Experiments_005700/005775/concat_csvs/"

length_single_axis.to_csv(os.path.join(results_dir + "measure_length_single_axis_segments_wapl_dep.csv"), index = False)
#average_length_wapl.to_csv(os.path.join(results_dir + "average_stats_resolved_dna_around_vermicelli_wapl_dep.csv"), index = False)
```

```python
sum_cohesin_axis =  length_cohesin_axis.groupby(['image_name'], as_index=False)[['length_cohesin_axis']].sum()
sum_cohesin_axis
```

```python
sum_single_axis =  length_single_axis.groupby(['image_name'], as_index=False)[['length_single_cohesin_axis']].sum()
sum_single_axis
```

```python
frames_merged = sum_single_axis.merge(sum_cohesin_axis, on='image_name',  how = "right")
fraction_single_axis = (frames_merged.loc[:, "length_single_cohesin_axis"]) / (frames_merged.loc[:, "length_cohesin_axis"])
percentage_single_axis = ((frames_merged.loc[:, "length_single_cohesin_axis"]) / (frames_merged.loc[:, "length_cohesin_axis"]))*100

frames_merged["fraction_single_axis"] = fraction_single_axis
frames_merged["percentage_single_axis"] = percentage_single_axis
```

```python
fraction_split_axes = 1 - (frames_merged.loc[:, "fraction_single_axis"])
```

```python
percentage_split_axes = fraction_split_axes *100
```

```python
frames_merged["fraction_split_axes"] = fraction_split_axes
frames_merged["percentage_split_axes"] = percentage_split_axes
```
