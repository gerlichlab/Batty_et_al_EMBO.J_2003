```python
path_split =  "/groups/gerlich/experiments/Experiments_005700/005765/quantify_fraction_split_axes/line_length_csv_files/wapl_sor_dep_0.75nm/0.75nm_split_axis_length/"


csv_files = glob.glob(path_split + "*.csv")

### Fetch all files in path
```

```python
all_df = []

#read in csv files
for file in csv_files:
    df = pd.read_csv(file, index_col=0, header=0)
    all_df.append(df)
    #change column names
    df.columns = ['image_name', 'area', 'mean', 'stddev', 'intden', 'rawintden', 'length']
    df.rename(columns = {'length':'length_split_cohesin_axis'}, inplace = True)
    df.rename(columns = {'label':'image_name'}, inplace = True)
    df["condition"] = "split_cohesin_axis_wapl_depleted"

#concat all dataframes together
frame = pd.concat(all_df, axis=0, ignore_index=True)
#remove duplicated rows
cols = frame.columns
#dedup = frame[cols].loc[(frame[cols].shift() != frame[cols]).any(axis=1)]

#slice relevant columns into final dataframe
length_split_axis = frame[['image_name','length_split_cohesin_axis', 'condition']]

#fully_prometa_two_sister_segment = fully_prometa_two_sister_segment.sort_values('edu_segment_length', ascending=False)
#fully_prometa_two_sister_segment_reset = fully_prometa_two_sister_segment.reset_index(drop=True)

results_dir = "/groups/gerlich/experiments/Experiments_005700/005765/quantify_fraction_split_axes/concat_csvs/"

length_split_axis.to_csv(os.path.join(results_dir + "measure_length_split_axis_segments_wapl_sor_dep_0.75nm.csv"), index = False)
#average_length_wapl.to_csv(os.path.join(results_dir + "average_stats_resolved_dna_around_vermicelli_wapl_dep.csv"), index = False)
```

```python
sum_split_axis =  length_split_axis.groupby(['image_name'], as_index=False)[['length_split_cohesin_axis']].sum()
sum_split_axis
```

```python
frames_merged_calculate_single = sum_single_axis.merge(sum_cohesin_axis, on='image_name',  how = "right")
fraction_single_axis = (frames_merged_calculate_single .loc[:, "length_single_cohesin_axis"]) / (frames_merged_calculate_single .loc[:, "length_cohesin_axis"])
percentage_single_axis = ((frames_merged_calculate_single .loc[:, "length_single_cohesin_axis"]) / (frames_merged_calculate_single.loc[:, "length_cohesin_axis"]))*100



frames_merged_calculate_single ["fraction_single_axis"] = fraction_single_axis
frames_merged_calculate_single ["percentage_single_axis"] = percentage_single_axis
```

```python
fraction_split_axes = 1 - (frames_merged_calculate_single.loc[:, "fraction_single_axis"])
percentage_split_axes = fraction_split_axes *100


frames_merged_calculate_single["fraction_split_axes"] = fraction_split_axes
frames_merged_calculate_single ["percentage_split_axes"] = percentage_split_axes
frames_merged_calculate_single.dropna()
```
