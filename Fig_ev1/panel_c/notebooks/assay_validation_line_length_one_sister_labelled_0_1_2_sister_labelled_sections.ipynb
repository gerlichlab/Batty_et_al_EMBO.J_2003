```python
path = "/groups/gerlich/experiments/Experiments_005600/005627/rois_hemi/one_sister_one_sister_length/"


csv_files = glob.glob(path + "*.csv")

### Fetch all files in path
```

```python
all_df = []

#read in csv files
for file in csv_files:
    df = pd.read_csv(file, index_col=0, header=0)
    all_df.append(df)
    #change column names
    df.columns = ['image_name', 'area', 'mean', 'stddev', 'intden', 'rawintden', 'length']
    df.rename(columns = {'length':'edu_segment_length'}, inplace = True)
    df.rename(columns = {'label':'image_name'}, inplace = True)
    df["condition"] = "hemi_prometa_one_sister_labelled_segment"

#concat all dataframes together
frame = pd.concat(all_df, axis=0, ignore_index=True)
#remove duplicated rows
cols = frame.columns
#dedup = frame[cols].loc[(frame[cols].shift() != frame[cols]).any(axis=1)]

#slice relevant columns into final dataframe
hemi_prometa_one_sister_segment = frame[['image_name','edu_segment_length', 'condition']]

hemi_prometa_one_sister_segment = hemi_prometa_one_sister_segment.sort_values('edu_segment_length', ascending=False)
hemi_prometa_one_sister_segment_reset = hemi_prometa_one_sister_segment.reset_index(drop=True)

results_dir = "/groups/gerlich/experiments/Experiments_005600/005627/results/"

hemi_prometa_one_sister_segment_reset.to_csv(os.path.join(results_dir + "measure_length_hemi_prometa_one_sister_labelled_segments.csv"), index = False)
#average_length_wapl.to_csv(os.path.join(results_dir + "average_stats_resolved_dna_around_vermicelli_wapl_dep.csv"), index = False)
```
